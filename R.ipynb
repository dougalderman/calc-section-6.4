```r
x<-c(1, 2, 4, 5)
y<-c(1, 3, 3, 4)
```

```r
plot(x,y,col="blue", main="y vs x",xlab="x coordinate", ylab="y coordinate")
```

```r
relation<-lm(y~I(x))
print(relation)
summary(relation)
```

```r
x_seq <- seq(from=1,to=5, by=0.01)
new_relation<-list(0)
for (i in 1:length(x_seq)) {
    x_new <- data.frame(x=x_seq[i])
    new_relation[i]<-predict(relation, x_new)
}
plot(x,y,col="blue", main="Linear Regression Example",lines(x_seq, new_relation),xlab="x",ylab="y")
```

```r
RSE_Percent = 0.7583/mean(y) * 100
```

```r
RSE_Percent
```

```r
confint(relation, level=0.95)
```

```r
plot(fitted(relation), residuals(relation))
```
